```python
# Lets plot Bar plot for Ahppiness score and countries
fig = plt.figure(figsize=(12,10))
plt.bar(data['Country'][0:10], data['Happiness.Score'][0:10])
```

```python
def HappyCountry(country,happyscore):
    happycontrylist = []
    happyscorelist = []
    for i,j in zip(country, happyscore):
        if(j>7):
            happycontrylist.append(i)
            happyscorelist.append(j)
    return happycontrylist,happyscorelist
        
    
happ_connt, happ_score = HappyCountry(data['Country'],data['Happiness.Score'])
```

```python
# Worlds most happy Countries
fig = plt.figure(figsize=(12,10))
plots = sns.barplot(happ_connt,happ_score)

for p in plots.patches:
    plots.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()))

plt.ylabel("Happiness Score")
plt.xlabel("Countries")

```

```python
# Lets have look at contries having less happiness score

def UnHappyCountry(country,happyscore):
    unhappycontrylist = []
    unhappyscorelist = []
    for i,j in zip(country, happyscore):
        if(j<4):
            unhappycontrylist.append(i)
            unhappyscorelist.append(j)
    return unhappycontrylist,unhappyscorelist

un_happy_country, un_happy_score = UnHappyCountry(data['Country'], data['Happiness.Score'])
```

```python
print(un_happy_country,"\n")
print(un_happy_score)
```

```python
len(un_happy_country)

# 20 conutries fall in this category where happiness-score is below 4
```

```python
# Lets plot the bar plot for Unhappy Countries, as per our consideration
fig = plt.figure(figsize=(12,10))
plots = sns.barplot(un_happy_country,un_happy_score)
# lets annotate
for p in plots.patches:
    plots.annotate(format(p.get_height(), '.2f'), (p.get_x()+p.get_width()/2,p.get_height()))

plt.ylabel("Happiness Score")
plt.xlabel("Countries")
```
